```python
import numpy as np
import matplotlib.pyplot as plt

def suma_serie_geometrica(a, r):
    if abs(r) < 1:
        return a / (1 - r)
    else:
        return np.inf  # no converge

def graficar_serie(a, r, nombre):
    n = np.arange(1, 20, 1)
    valores = a * r**(n - 1)
    suma_parcial = np.cumsum(valores)

    plt.figure()
    plt.plot(n, suma_parcial, 'o-', label=f'suma parcial de {nombre}')
    if abs(r) < 1:
        plt.axhline(y=a / (1 - r), color='r', linestyle='--', label='asintota')

    plt.xlabel('n')
    plt.ylabel('suma parcial')
    plt.legend()
    plt.title(f'convergencia de {nombre}')
    plt.grid()
    plt.show()

# lista de series y sus valores
series = [
    (1, 1/3, 'serie a'),
    (4, 0.21, 'serie b'),
    (2, 1/5, 'serie c'),
    (2, 1/7, 'serie d'),
    (7, (1/3) * (2/5), 'serie e'),
    (1, 2/3, 'serie f'),
    (1, 3/2, 'serie g'),  # no converge
    (1, 0.9999, 'serie h'),
    (1, 1.00001, 'serie i'),  # no converge
    (5, 3/7, 'serie j'),
    (1, 91/89, 'serie k'),  # no converge
    (1, np.pi/4, 'serie l'),  # no converge
]

# calcular y graficar cada serie
for a, r, nombre in series:
    suma = suma_serie_geometrica(a, r)
    print(f'{nombre}: {suma if suma != np.inf else "diverge"}')
    if suma != np.inf:
        graficar_serie(a, r, nombre)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

def factorial_series(n_terminos):
    terminos = np.array([1 / factorial(n) for n in range(n_terminos)])
    sumas_parciales = np.cumsum(terminos)
    return terminos, sumas_parciales

n_terminos = 20

terminos, s_n = factorial_series(n_terminos)

# demostrar que s_n < 3 para todo n
es_menor_que_3 = all(s_n < 3)
print(f"para todo n, s_n < 3: {es_menor_que_3}")

# determinar cuntos trminos se necesitan para que la suma sea menor que 0.000001 de error con e
error_tolerado = 0.000001
n_requerido = np.argmax(np.abs(s_n - np.e) < error_tolerado) + 1
print(f"se necesitan {n_requerido} terminos para estar dentro de {error_tolerado} de e")

# graficar la serie
plt.plot(range(1, n_terminos+1), s_n, marker='o', linestyle='-', label='suma parcial')
plt.axhline(y=np.e, color='r', linestyle='--', label='asintota (e)')
plt.axhline(y=3, color='g', linestyle='--', label='cota superior 3')
plt.xlabel('numero de terminos')
plt.ylabel('suma parcial')
plt.title('convergencia de la serie de factoriales')
plt.legend()
plt.grid()
plt.show()

print(f"valor aproximado de la serie con {n_terminos} terminos: {s_n[-1]}")
print(f"valor real de e: {np.e}")
```

```python
# calcular cuntos trminos de la serie armnica se necesitan para superar un lmite

def encontrar_n_para_suma_limite(limite):
    suma = 0
    n = 0
    while suma <= limite:
        n += 1
        suma += 1 / n
    return n

# calcular para h_n > 5 y h_n > 10
n_5 = encontrar_n_para_suma_limite(5)
n_10 = encontrar_n_para_suma_limite(10)

print("numero de terminos para h_n > 5:", n_5)
print("numero de terminos para h_n > 10:", n_10)
```

```python
# funcion para calcular la distancia total recorrida por la pelota
def distancia_total(altura_inicial, razon):
    caida_inicial = altura_inicial
    suma_rebotes = (2 * altura_inicial * razon) / (1 - razon)
    return caida_inicial + suma_rebotes

# caso 1: rebota al 75% de la altura anterior
distancia_75 = distancia_total(10, 0.75)

# caso 2: rebota al 25% de la altura anterior
distancia_25 = distancia_total(10, 0.25)

print("distancia total si rebota al 75%:", distancia_75, "metros")
print("distancia total si rebota al 25%:", distancia_25, "metros")
```
