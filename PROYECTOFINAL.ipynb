Resumen ejecutivo

Este proyecto analiza el crecimiento de usuarios de la aplicación estudiantil “StudyBoost” durante sus primeras ocho semanas en el mercado, bajo dos estrategias de marketing digital: micro influencers en redes sociales (Estrategia A) y publicidad paga en Google Ads (Estrategia B). Para modelar y comparar el comportamiento del crecimiento, se utilizaron conceptos matemáticos como sucesiones aritméticas y geométricas. Se simularon datos realistas y se contrastaron con el caso de una app real (Forest), reconocida en el ámbito académico. La metodología incluyó la construcción de modelos matemáticos, análisis de tasas de crecimiento y comparación gráfica de resultados. Los hallazgos indican que, si bien la estrategia de publicidad paga genera un crecimiento más rápido al inicio, el uso de micro influencers puede alcanzar tasas de crecimiento similares o superiores a largo plazo. Esto sugiere que una combinación de ambas estrategias podría ser la más efectiva para el posicionamiento sostenible de la aplicación.

Introducción

En el desarrollo y lanzamiento de aplicaciones móviles, una de las principales preocupaciones de los desarrolladores es lograr un crecimiento sostenido en la cantidad de usuarios. En el caso de StudyBoost, una app de recordatorios académicos orientada a estudiantes universitarios, se planteó el reto de evaluar cuál estrategia de marketing digital resulta más efectiva en sus primeras semanas en el mercado: el uso de micro influencers en redes sociales o una campaña paga en Google Ads.

Este análisis es relevante para el campo de la Ingeniería de Sistemas, ya que combina el desarrollo de software con la toma de decisiones basada en datos, permitiendo optimizar recursos en fases clave del producto.

Las matemáticas aplicadas, en especial el cálculo y el análisis de funciones, permiten modelar el crecimiento de usuarios mediante sucesiones y funciones matemáticas. Esto facilita comparar escenarios, prever resultados futuros y tomar decisiones informadas desde una perspectiva cuantitativa y objetiva.



Marco teórico

Una sucesión aritmética es una secuencia numérica donde cada término se obtiene sumando una cantidad constante al anterior. Esta estructura fue utilizada para representar el comportamiento de la Estrategia B (campaña paga en Google Ads), dado que se espera un crecimiento constante semanal producto de una inversión regular en publicidad.

Por otro lado, una sucesión geométrica es una secuencia en la cual cada término se obtiene multiplicando el anterior por una razón fija. Este modelo se aplicó en la Estrategia A (publicidad con micro influencers), donde se simula un crecimiento más lento al inicio pero que se acelera progresivamente por efecto del boca a boca y las recomendaciones en redes sociales.

El uso de estos modelos matemáticos permite representar de forma precisa el comportamiento del número de usuarios a lo largo del tiempo y facilita la comparación entre ambas estrategias, apoyando la toma de decisiones basada en datos.

Metodología

Para analizar el crecimiento de usuarios en StudyBoost, se definieron dos escenarios simulados que representan estrategias distintas de marketing digital:

Estrategia A: Publicidad en redes sociales mediante micro influencers, con un crecimiento acelerado modelado como una sucesión geométrica.

Estrategia B: Campaña paga en Google Ads, representada por una sucesión aritmética con crecimiento constante semanal.

Se construyeron tablas de datos para 8 semanas, partiendo de un número inicial de usuarios y aplicando las fórmulas respectivas para cada tipo de sucesión. Posteriormente, se compararon estos resultados con los datos estimados de una aplicación real del mismo sector (Forest: Stay Focused), para verificar la coherencia de los escenarios simulados.

Procesos matemáticos aplicados:
Modelación con sucesiones: se usaron fórmulas explícitas para calcular el número de usuarios semana a semana.

Cálculo de tasas de crecimiento: se analizó el incremento semanal para comparar la efectividad de cada estrategia.

Comparación de escenarios: se graficaron y analizaron los datos de las dos estrategias frente a los de la app real.

Supuestos realizados:
Cada estrategia se aplica de forma exclusiva durante las 8 semanas.

No hay desinstalaciones ni usuarios inactivos.

El crecimiento responde únicamente a la estrategia de marketing y no a otros factores externos como mejoras en la app.

Los usuarios son nuevos (no hay registros duplicados).

Los datos de la aplicación real fueron estimados con base en fuentes públicas no oficiales, con fines comparativos.

Resultados

a=200

t= numero de semanas



\begin{aligned}

Sucesión arimetica (b)

a=800 d=600 t=numero de semanas

\begin{aligned}
U_B(0) &= 800 + 600 \cdot 0 = 800 \\

```python
import matplotlib.pyplot as plt
import pandas as pd

# semanas
semanas = list(range(9))  # De la semana 0 a la 8

# estrategia A: sucesion geometrica
a_geom = 200
r = 1.5
usuarios_A = [round(a_geom * (r ** t)) for t in semanas]

# estrategia B: sucesion aritmetica
a_arit = 800
d = 600
usuarios_B = [a_arit + d * t for t in semanas]

# app real (forest) – valores estimados
usuarios_Forest = [500, 850, 1400, 2300, 3600, 5000, 7000, 9500, 12000]

# crear DataFrame para visualizar en tabla
df = pd.DataFrame({
    'semana': semanas,
    'estrategia A (geometrica)': usuarios_A,
    'estrategia B (aritmetica)': usuarios_B,
    'forest (referente)': usuarios_Forest
})

# mostrar tabla
print( "tabla de usuarios por semana:\n")
print(df.to_string(index=False))

# graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(semanas, usuarios_A, marker='o', label='estrategia A (microinfluencers)')
plt.plot(semanas, usuarios_B, marker='s', label='estrategia B (publicidad paga)')
plt.plot(semanas, usuarios_Forest, marker='^', label='forest (app real de referente)', linestyle='--')

plt.title('comparación de crecimiento de usuarios')
plt.xlabel('semana')
plt.ylabel('numero de usuarios')
plt.grid(True)
plt.legend()
plt.xticks(semanas)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# datos de la estrategia a
semanas = list(range(9))
usuarios_a = [round(200 * (1.5) ** t) for t in semanas]

# grafica
plt.figure(figsize=(8, 5))
plt.plot(semanas, usuarios_a, marker='o', color='blue', label='estrategia a (geometrica)')
plt.title('crecimiento de usuarios - estrategia a')
plt.xlabel('semana')
plt.ylabel('numero de usuarios')
plt.grid(True)
plt.xticks(semanas)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# datos de la estrategia b
semanas = list(range(9))
usuarios_b = [800 + 600 * t for t in semanas]

# grafica
plt.figure(figsize=(8, 5))
plt.plot(semanas, usuarios_b, marker='s', color='green', label='estrategia b (aritmetica)')
plt.title('crecimiento de usuarios - estrategia b')
plt.xlabel('semana')
plt.ylabel('numero de usuarios')
plt.grid(True)
plt.xticks(semanas)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# datos estimados de forest
semanas = list(range(9))
usuarios_forest = [500, 850, 1400, 2300, 3600, 5000, 7000, 9500, 12000]

# grafica
plt.figure(figsize=(8, 5))
plt.plot(semanas, usuarios_forest, marker='^', color='orange', linestyle='--', label='forest (app real)')
plt.title('crecimiento de usuarios - app forest')
plt.xlabel('semana')
plt.ylabel('numero de usuarios')
plt.grid(True)
plt.xticks(semanas)
plt.legend()
plt.show()
```
