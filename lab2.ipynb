punto 1

La funcion es $f(x) = \frac{x}{x+1}$.


- **dominio**: todos los reales excepto \( x = -1 \)

- **simetria**: no es par ni impar

- **intersecciones**:  
  - eje x: \( x = 0 \)  
  - eje y: \( y = 0 \)

- **asintotas**:  
  - vertical: \( x = -1 \)  
  - horizontal: \( y = 1 \)

```python
from matplotlib import pyplot as plt
import numpy as np

# funcion 1: f(x) = x / (x + 1)
x = np.linspace(-10, 10, 400)
x = x[x != -1]  # evitar division por cero
y = x / (x + 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x / (x + 1)', color='orange')
plt.axhline(1, color='gray', linestyle='--', label='asintota horizontal y = 1')
plt.axvline(-1, color='red', linestyle='--', label='asintota vertical x = -1')
plt.title('grafica de f(x) = x / (x + 1)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

punto 3 $$ f(x) = \frac{x^2}{x^2 - 1} $$

 Anlisis de la funcin

#### (a) Dominio y Rango


$$ x^2 - 1 = 0 \Rightarrow x^2 = 1 \Rightarrow x = \pm 1 $$



$$ x \in \mathbb{R} \setminus \{-1,1\} $$

El rango de la funcin es \( y \neq 1 \), ya que no hay ningn valor de \( x \) que haga que \( f(x) = 1 \).

#### (b) Simetra
Para verificar la paridad de la funcin, evaluamos \( f(-x) \):

$$ f(-x) = \frac{(-x)^2}{(-x)^2 - 1} = \frac{x^2}{x^2 - 1} = f(x) $$

Como \( f(-x) = f(x) \), la funcin es **par**, lo que significa que es simtrica respecto al eje \( y \).

#### (c) Intersecciones y Asntotas
- **Interseccin con el eje \( x \)**: Se anula cuando \( f(x) = 0 \), lo que ocurre en \( x = 0 \). La interseccin es \( (0,0) \).
- **Interseccin con el eje \( y \)**: Se encuentra en \( x = 0 \), con \( f(0) = 0 \). La interseccin es \( (0,0) \).
- **Asntotas Verticales**: La funcin no est definida en \( x = \pm 1 \), lo que indica **asntotas verticales** en \( x = -1 \) y \( x = 1 \).
- **Asntota Horizontal**: Para \( |x| \to \infty \), la funcin tiende a:

$$ \lim_{x \to \pm \infty} f(x) = 1 $$

Esto indica una **asntota horizontal** en \( y = 1 \).

lab.jpg

```python
import numpy as np
import matplotlib.pyplot as plt

# definir la funcion
def f(x):
    return x**2 / (x**2 - 1)

# crear valores de x evitando x = 1
x = np.linspace(-3, 3, 400)
x = x[x != -1]
x = x[x != 1]

# calcular valores de y
y = f(x)

# graficar la funcion
plt.plot(x, y, label="f(x) = x^2 / (x^2 - 1)")
plt.axhline(1, color="r", linestyle="--", label="asintota y = 1")
plt.axvline(-1, color="g", linestyle="--", label="asintota x = -1")
plt.axvline(1, color="g", linestyle="--", label="asintota x = 1")

# formato de la grafica
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("grafica de la funcion f(x)")
plt.legend()
plt.grid()

# mostrar la grafica
plt.show()
```

punto 2  $g(x) = \frac{|x| + x}{2}(3x + 2)$

Dominio y Rango

Dominio: $\mathbb{R}$
Rango: $[0, +\infty)$

 Simetra

No es par ni impar

 Intersecciones y Asntotas

Interseccin con eje y: $(0, 0)$
Intersecciones con eje x: Todo el semieje $(-\infty, 0]$
Asntotas: Horizontal $y = 0$ para $x \to -\infty$

```python
import numpy as np
import matplotlib.pyplot as plt

# definir la funcion
def g(x):
    return np.where(x >= 0, x * (3*x + 2), 0)

# crear valores de x
x = np.linspace(-3, 3, 1000)
y = g(x)

# crear la grafica
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2, label='g(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# etiquetas y titulo
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('grafica de g(x) = (|x| + x)/2 * (3x + 2)')
plt.legend()

# mostrar puntos importantes
plt.plot(0, 0, 'ro', markersize=8, label='interseccion (0,0)')
plt.legend()

plt.show()
```

punto 4 $h(x) = \frac{5.235x^3 - 1.245}{7.126x - 3.141}$

Dominio y Rango

Dominio: $\mathbb{R} - {\frac{3.141}{7.126}} = \mathbb{R} - {0.441}$
Rango: $\mathbb{R}$

Simetra

No es par ni impar

 Intersecciones y Asntotas

Interseccin con eje y: $(0, 0.396)$
Intersecciones con eje x: $(0.694, 0)$
Asntota vertical: $x = 0.441$
Asntota oblicua: $y = 0.735x^2 + 0.324x + 0.143$


 Anlisis con Primera Derivada

Puntos crticos: $x = -0.295$ y $x = 1.177$
Funcin creciente: $(-\infty, -0.295) \cup (1.177, \infty)$
Funcin decreciente: $(-0.295, 0.441) \cup (0.441, 1.177)$

 Mximos y Mnimos Locales

Mximo local: $(-0.295, 1.203)$
Mnimo local: $(1.177, -0.411)$

 Anlisis con Segunda Derivada

Cncava hacia arriba: $(0.441, \infty)$
Cncava hacia abajo: $(-\infty, 0.441)$
Puntos de inflexin: Ninguno (discontinuidad en $x = 0.441$)

```python
import numpy as np
import matplotlib.pyplot as plt

# definir la funcion
def h(x):
    return (5.235*x**3 - 1.245) / (7.126*x - 3.141)

# crear valores de x evitando la asintota vertical en x = 0.441
x1 = np.linspace(-3, 0.44, 500)
x2 = np.linspace(0.442, 3, 500)

y1 = h(x1)
y2 = h(x2)

# crear la grafica
plt.figure(figsize=(10, 8))

# graficar las dos partes de la funcion
plt.plot(x1, y1, 'b-', linewidth=2, label='h(x)')
plt.plot(x2, y2, 'b-', linewidth=2)

# asintota vertical
plt.axvline(x=0.441, color='r', linestyle='--', alpha=0.7, label='asintota vertical x=0.441')

# puntos importantes
plt.plot(0, 0.396, 'ro', markersize=8, label='interseccion y (0, 0.396)')
plt.plot(0.694, 0, 'go', markersize=8, label='interseccion x (0.694, 0)')
plt.plot(-0.295, 1.203, 'mo', markersize=8, label='maximo local (-0.295, 1.203)')
plt.plot(1.177, -0.411, 'co', markersize=8, label='minimo local (1.177, -0.411)')

# configuracion de la grafica
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('grafica de h(x) = (5.235x^3 - 1.245)/(7.126x - 3.141)')
plt.legend()
plt.ylim(-5, 5)
plt.xlim(-3, 3)

plt.show()
```
