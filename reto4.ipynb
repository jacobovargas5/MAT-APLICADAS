```python
import matplotlib.pyplot as plt

def fibonacci(n):
    series = [0, 1]
    for i in range(2, n):
        series.append(series[-1] + series[-2])
    return series[:n]

def plot_fibonacci(n):
    series = fibonacci(n)
    plt.plot(series, marker='o', linestyle='-')
    plt.xlabel('Índice')
    plt.ylabel('Valor')
    plt.title('Serie de Fibonacci')


n = int(input("Ingrese un número: "))
plot_fibonacci(n)
```

```python
import matplotlib.pyplot as plt

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
                26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = sum(temperaturas) / len(temperaturas)
mediana = sorted(temperaturas)[len(temperaturas) // 2]
desviacion = (sum((x - media) ** 2 for x in temperaturas) / len(temperaturas)) ** 0.5

print(f"media: {media:.2f}")
print(f"mediana: {mediana:.2f}")
print(f"desviacion estandar: {desviacion:.2f}")

plt.plot(temperaturas, marker='o')
plt.xlabel('dias')
plt.ylabel('temperatura (°C)')
plt.title('temperaturas diarias')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Punto 5: Raíces cuadradas
def graficar_raices():
    x = np.linspace(0, 10, 400)
    y = np.sqrt(x)

    plt.figure()
    plt.plot(x, y, label='y = √x', color='b')
    plt.plot(x, x, label='y = x', color='r', linestyle='--')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Raíz cuadrada y función identidad')
    plt.grid()
    plt.show()

graficar_raices()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
# Punto 6: Pendiente de una recta
def graficar_recta(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1

    x = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 100)
    y = m * x + b

    plt.figure()
    plt.plot(x, y, label=f'Recta: y = {m:.2f}x + {b:.2f}', color='g')
    plt.scatter([x1, x2], [y1, y2], color='r', label='Puntos dados')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Pendiente de una recta')
    plt.grid()
    plt.show()

graficar_recta(1, 2, 4, 6)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
# Punto 7: Movimiento parabólico
def movimiento_parabolico(velocidad_inicial, angulo):
    g = 9.81
    angulo_rad = np.radians(angulo)
    t_vuelo = (2 * velocidad_inicial * np.sin(angulo_rad)) / g
    t = np.linspace(0, t_vuelo, num=100)
    x = velocidad_inicial * np.cos(angulo_rad) * t
    y = velocidad_inicial * np.sin(angulo_rad) * t - (0.5 * g * t**2)

    plt.figure()
    plt.plot(x, y, label=f'Ángulo: {angulo}°', color='b')
    plt.xlabel('Distancia (m)')
    plt.ylabel('Altura (m)')
    plt.title('Movimiento parabólico')
    plt.legend()
    plt.grid()
    plt.show()

movimiento_parabolico(20, 45)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
# Punto 8: Modificación de función con un deslizador
def actualizar(val):
    n = int(slider.val)
    y = x ** n
    linea.set_ydata(y)
    fig.canvas.draw_idle()

x = np.linspace(1, 10, 400)
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)
linea, = plt.plot(x, x ** 1, label='x^n', color='m')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funciones de la forma x^n')
plt.legend()
plt.grid()
ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])
slider = Slider(ax_slider, 'n', 1, 10, valinit=1, valstep=1)
slider.on_changed(actualizar)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Punto 9: Fractal de Sierpinski
def sierpinski(puntos, iteraciones):
    triangulo = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
    x, y = [0.5], [0.5]

    for _ in range(iteraciones):
        vertice = triangulo[np.random.randint(0, 3)]
        x.append((x[-1] + vertice[0]) / 2)
        y.append((y[-1] + vertice[1]) / 2)

    plt.figure()
    plt.scatter(x, y, s=1, color='purple')
    plt.title('Fractal de Sierpinski')
    plt.axis('equal')
    plt.show()

sierpinski(5000, 5000)
```
